```python
import random as r
num_set = set([-12,-3,-6,7,2,-2,6,3,9,-7,-5,-8,1,11,-9,-4])

US = 6
LS = 3
ResultList = set()
Iterations = 1000

for i in range (Iterations):
  for k in range (LS,US+1):
    Chromosome = r.sample(list(num_set),k)

    if sum (Chromosome)==0:
     ResultList.add(tuple(Chromosome))
     Chromosome.sort()

#used nested loops  in this question to iterate the size of set


for r in ResultList:
  print(r)





print(len(ResultList))
```
